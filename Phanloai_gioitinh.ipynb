```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import string
```

```python
df1=pd.read_csv("/content/female.tsv", names=["Name","Gender"], sep="\t")
df2=pd.read_csv("/content/male.tsv", names=["Name","Gender"], sep="\t")
```

```python
df1['Gender']=0 #0 là giới tính nữ
df2['Gender']=1 #1 là giới tính nam
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction import DictVectorizer
```

```python
names = df3["Name"]
```

```python
names
```

```python
names[0]
```

```python
cv = CountVectorizer()
X = cv.fit_transform(names.values.astype('U'))
```

```python
cv.get_feature_names()
```

```python
X
y = df3.Gender
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.5, random_state=0)
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
print("Độ chính xác của mô hình",model.score(X_test, y_test)*100,"%")
```
